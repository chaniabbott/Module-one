copying tuples is unnecessary because you can just point to its place

```python
tt = (1,2,['a','b'])
uu = tt
print(f'{uu is tt}')
tt[2].append('c')
print(f'{tt=} {uu=}')
```

```python
import copy #this works for lists also

uu = copy.deepcopy(tt)
print(f'{uu is tt}')
tt[2].append('d')
print(f'{tt=} {uu=}')
```
